```python
# Data Exploration customers_df
customers_df.describe(include="all")
```

```python
customers_df.groupby(by="gender").agg({ 
    "customer_id": "nunique", 
    "age": ["max", "min", "mean", "std"]
})
```

```python
customers_df.groupby(by="city").customer_id.nunique().sort_values(ascending=False)
customers_df.groupby(by="state").customer_id.nunique().sort_values(ascending=False)
```

```python
# Data Exploration orders_df
delivery_time = orders_df["delivery_date"] - orders_df["order_date"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
orders_df["delivery_time"] = round(delivery_time/86400)
```

```python
orders_df.describe(include="all")
```

```python
# Data Exploration orders_df & customers_df
customer_id_in_orders_df = orders_df.customer_id.tolist()
customers_df["status"] = customers_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_orders_df else "Non Active")
customers_df.sample(5)
```

```python
customers_df.groupby(by="status").customer_id.count()
```

```python
orders_customers_df = pd.merge(
    left = orders_df, 
    right = customers_df, 
    how = "left", 
    left_on = "customer_id", 
    right_on = "customer_id"
)

orders_customers_df.head()
```

```python
# Amount of order based city
orders_customers_df.groupby(by="city").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

```python
# Amount of order based state
orders_customers_df.groupby(by="state").order_id.nunique().sort_values(ascending=False)
```

```python
# Amount of order based gender
orders_customers_df.groupby(by="gender").order_id.nunique().sort_values(ascending=False)
```

```python
# Amount of order based age group
orders_customers_df["age_group"] = orders_customers_df.age.apply(lambda x: "Youth" if x <= 24 else ("Seniors" if x > 64 else "Adults"))
orders_customers_df.groupby(by="age_group").order_id.nunique().sort_values(ascending=False)
```

```python
# Data Exploration products_df & sales_df 

# products_df
products_df.describe(include="all")
```

```python
# sales_df
sales_df.describe(include="all")
```

```python
# Sort values price of products_df
products_df.sort_values(by="price", ascending=False)
```

```python
# Use pivot table to gather the information of products based type and name
products_df.groupby(by="product_type").agg({ 
    "product_id": "nunique", 
    "quantity": "sum", 
    "price": ["min", "max"]
})
```

```python
products_df.groupby(by="product_name").agg({ 
    "product_id": "nunique", 
    "quantity": "sum", 
    "price": ["min", "max"]
})
```

```python
# Merge products_df and sales_df to know best selling products
sales_products_df = pd.merge(
    left=sales_df, 
    right=products_df, 
    how="left",
    left_on="product_id", 
    right_on="product_id"
)

sales_products_df.head()
```

```python
# View the information of products selling based type
sales_products_df.groupby(by="product_type").agg({
    "sales_id": "nunique", 
    "quantity_x": "sum", 
    "total_price": "sum"
})
```

```python
# View the information of products selling based name
sales_products_df.groupby(by="product_name").agg({ 
    "sales_id": "nunique", 
    "quantity_x": "sum", 
    "total_price": "sum"
}).sort_values(by="total_price", ascending=False)
```

```python
# Data Exploration all_df
# Merge 4 tables to collect all the information from those tables
all_df = pd.merge(
    left=sales_products_df,
    right=orders_customers_df,
    how="left",
    left_on="product_id",
    right_on="order_id"
)

all_df.head()
```

```python
# View the preference of purchases based on customers state and product type
all_df.groupby(by=["state", "product_type"]).agg({
    "quantity_x": "sum", 
    "total_price": "sum"
})
```

```python
# View the preference of purchases based on customers gender and product type
all_df.groupby(by=["gender", "product_type"]).agg({
    "quantity_x": "sum",
    "total_price": "sum"
})
```

```python
# View the preference of purchases based on customers age group and product type
all_df.groupby(by=["age_group", "product_type"]).agg({
    "quantity_x": "sum",
    "total_price": "sum"
})
```
